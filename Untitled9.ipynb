```python
import pandas as pd

# STEP 1: Load the CSV file
df = pd.read_csv("superstore.csv", encoding="latin-1")

# STEP 2: First look at the data
print("Shape:", df.shape)

print("\nFirst 5 rows:")
print(df.head())

print("\nColumn names:")
print(df.columns.tolist())

print("\nMissing values:")
print(df.isnull().sum())

# STEP 3: Clean column names
df.columns = (
    df.columns
    .str.lower()
    .str.replace(" ", "_")
    .str.replace("-", "_")
)

print("\nCleaned columns:")
print(df.columns.tolist())

# STEP 4: Fix data types
df["order_date"] = pd.to_datetime(df["order_date"], format="%d/%m/%Y")
df["ship_date"] = pd.to_datetime(df["ship_date"], format="%d/%m/%Y")

# STEP 5: Drop rows with missing important values
# The 'profit' column does not exist in the dataset, so removing it from dropna subset
df.dropna(subset=["sales", "order_id"], inplace=True)

# STEP 6: Add profit margin column (skipped as 'profit' column is missing)
# If you intend to have a 'profit' column, please ensure it's present in your CSV.
# df["profit_margin"] = (df["profit"] / df["sales"]).round(4)

# STEP 7: Final check
print("\nCleaned data shape:", df.shape)
print(df.dtypes)

# Save cleaned file
df.to_csv("superstore_cleaned.csv", index=False)

print("\nCleaned file saved!")
```
